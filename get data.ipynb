```python
import json
import numpy as np
import pandas as pd
import requests
import time
```

```python
URL = "https://opendata.sz.gov.cn/api/29200_01003608/1/service.xhtml"

HEADERS = {
    "User-Agent": "Mozilla/5.0",
    "Referer": "https://opendata.sz.gov.cn/"
}

APP_KEY = "MY_API_KEY" 


def pull_page(page):
    params = {
        "page": page,
        "rows": 1000,   
        "appKey": APP_KEY
    }

    response = requests.post(URL, data=params, headers=HEADERS)
    response.raise_for_status()

    return response.json()


def pull_all():
    all_rows = []
    page = 1

    while True:
        print(f"正在抓取第 {page} 页...")
        
        data = pull_page(page)
        rows = data.get("data", [])

        if not rows:
            print("没有更多数据，停止抓取")
            break

        all_rows.extend(rows)

        page += 1
        time.sleep(0.3)  # 防止被限流

    return all_rows
```

```python
data = pull_all()
print("总数据条数:", len(data))
```

```python
air_quality_df = pd.DataFrame(data)
print(air_quality_df.head())
air_quality_df.to_csv( "../data/air_quality_data.csv", index=False)
```
