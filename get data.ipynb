```python
import json
import numpy as np
import pandas as pd
import requests
import time
```

```python
URL = "https://opendata.sz.gov.cn/api/29200_01003608/1/service.xhtml"

HEADERS = {
    "User-Agent": "Mozilla/5.0",
    "Referer": "https://opendata.sz.gov.cn/"
}

APP_KEY = "MY_API_KEY" 


def pull_page(page):
    params = {
        "page": page,
        "rows": 1000,   
        "appKey": APP_KEY
    }

    response = requests.post(URL, data=params, headers=HEADERS)
    response.raise_for_status()

    return response.json()


def pull_all():
    all_rows = []
    page = 1

    while True:
        print(f"正在抓取第 {page} 页...")
        
        data = pull_page(page)
        rows = data.get("data", [])

        if not rows:
            print("没有更多数据，停止抓取")
            break

        all_rows.extend(rows)

        page += 1
        time.sleep(0.3)  # 防止被限流

    return all_rows
```

```python
data = pull_all()
print("总数据条数:", len(data))
```

```python
air_quality_df = pd.DataFrame(data)
print(air_quality_df.head())
air_quality_df.to_csv( "../data/air_quality_data.csv", index=False)
```

由于下载速度过慢，使用一个带断点检测的脚本下载数据

```python
import cdsapi
import calendar
from datetime import date
from pathlib import Path

dataset = "reanalysis-era5-single-levels"

variables = [
    "10m_u_component_of_wind",
    "10m_v_component_of_wind",
    "2m_dewpoint_temperature",
    "2m_temperature",
    "surface_pressure",
    "total_precipitation",
    "boundary_layer_height",
]

times = [f"{h:02d}:00" for h in range(24)]
area = [22.9, 113.7, 22.4, 114.7]

start_year = 2020
end_date = date.today()

out_dir = Path("../data/era5").resolve()
out_dir.mkdir(parents=True, exist_ok=True)
client = cdsapi.Client()

for year in range(start_year, end_date.year + 1):
    for month in range(1, 13):

        if year == end_date.year and month > end_date.month:
            break

        last_day = calendar.monthrange(year, month)[1]
        if year == end_date.year and month == end_date.month:
            last_day = min(last_day, end_date.day)

        target = out_dir / f"era5_single_{year}{month:02d}.zip"

        # 如果文件存在就跳过
        if target.exists():
            print(f"Skip {target.name} (already exists)")
            continue

        request = {
            "product_type": ["reanalysis"],
            "variable": variables,
            "year": [str(year)],
            "month": [f"{month:02d}"],
            "day": [f"{d:02d}" for d in range(1, last_day + 1)],
            "time": times,
            "data_format": "netcdf",
            "download_format": "unarchived",
            "area": area,
        }

        print("Downloading:", target.name)
        client.retrieve(dataset, request).download(str(target))
```
